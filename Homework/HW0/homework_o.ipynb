```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the data from file
relative_path = 'iris/bezdekIris.data' 
with open(relative_path, 'r') as file:
    lines = file.readlines()
```

```python
# Definition of the array 
sepal_length = []
sepal_width = []
petal_length = []
petal_width = []
class_flower = []

# Split the data
for line in lines:
    parts = line.strip().split(',')
    sepal_length.append(float(parts[0]))
    sepal_width.append(float(parts[1]))
    petal_length.append(float(parts[2]))
    petal_width.append(float(parts[3]))
    class_flower.append(parts[4])


# Transfer the list to np array
sepal_length = np.array(sepal_length)
sepal_width = np.array(sepal_width)
petal_length = np.array(petal_length)
petal_width = np.array(petal_width)
```

```python
# Prepare the array needed
data_arrays = [sepal_length, sepal_width, petal_length, petal_width]
labels = ['Sepal length', 'Sepal width', 'Petal length', 'Petal width']
num_arrays = len(data_arrays)

#Set colors
colors = []
for k in range(50):
    colors.append('red')
for k in range(50):
    colors.append('blue')
for k in range(50):
    colors.append('green')

# Draw the plot
for i in range(num_arrays):
    for j in range(i+1, num_arrays):
        plt.scatter(data_arrays[i], data_arrays[j], c=colors, 
                    label=f"{labels[i]} vs {labels[j]}")
        plt.xlabel(f"{labels[i]}")
        plt.ylabel(f"{labels[j]}")
        plt.grid(True)
        plt.title(f"{labels[i]} vs {labels[j]} Scatter Plot") 
        plt.savefig(f"{i}{j}.png")
        plt.show()

for i in range(num_arrays):
    for j in range(i+1, num_arrays):
        plt.scatter(data_arrays[j], data_arrays[i], c=colors, 
                    label=f"{labels[j]} vs {labels[i]}")
        plt.xlabel(f"{labels[j]}")
        plt.ylabel(f"{labels[i]}")
        plt.grid(True)
        plt.title(f"{labels[i]} vs {labels[j]} Scatter Plot") 
        plt.savefig(f"{j}{i}.png")
        plt.show()
        
```
